```python
def is_cyclic(graph_array):
    used = [False] * len(graph_array)

    def dfs(v, p=-1):
        used[v] = True
        for u in graph_array[v]:
            if not used[u]:
                if dfs(u, v):
                    return True
            elif u != p:
                return True
        return False

    for i in range(len(graph_array)):
        if not used[i]:
            if dfs(i):
                return True
    return False


V, E = map(int, input().split())
a = input()[1:-1]

gr = []
for j in range(V):
    gr.append(list(map(int, a[a.index('{') + 1:a.index('}')].split(', '))))
    a = a[a.index('}') + 1:]
print(is_cyclic(gr))
```

```python
def max_meetings(start, end, n_ver):
    visited = [False] * n_ver
    graph = {i: [] for i in range(n_ver)}
    for i in range(n_ver):
        for j in range(n_ver):
            if end[i] < start[j]:
                graph[i].append(j)

    def dfs(v):
        if visited[v]:
            return visited[v]
        max_len = 1
        for u in graph[v]:
            max_len = max(max_len, 1 + dfs(u))
        visited[v] = max_len
        return max_len

    res = 0
    for i in range(n_ver):
        res = max(res, dfs(i))
    return res


n = int(input())
s = list(map(int, input().split()))
e = list(map(int, input().split()))

print(max_meetings(s, e, n))
```

```python
import sys

V, E, start, end = map(str, input().split())
V, E = int(V), int(E)
gr = [list(map(str, input().split())) for _ in range(E)]
graf = {}
for i in range(E):
    graf[gr[i][0]] = dict()
    graf[gr[i][1]] = dict()
for i in range(E):
    graf[gr[i][0]] |= {gr[i][1]: int(gr[i][2])}


# graf = {'A': {'B': 1, 'D': 1, 'C': 3},
#         'B': {'A': 1, 'E': 5},
#         'C': {'A': 3, 'D': 1},
#         'D': {'A': 1, 'C': 1, 'E': 3, 'F': 9},
#         'E': {'B': 5, 'D': 3, 'G': 7},
#         'G': {'E': 7, 'F': 8},
#         'F': {'D': 9, 'G': 8}
#         }


# возвращает красивым списочком все вершины графа
def all_vertex(graf):
    return list(graf.keys())


# возвращаем не менее красивым списочком всех соседей данной вершины
def vertex_neighbors(graf, vertex):
    return list(graf[vertex].keys())


# возвращаем длину ребра между двумя вершинами
def value(graf, vertex1, vertex2):
    return (graf[vertex1][vertex2])


def dijkstra(graf, start):
    unvisited_vertexes = all_vertex(graf)
    shortest_path = {}
    previous_vertex = {}

    # по умолчанию расстояния между любыми двумя вершинами задаем равным бесконечности,
    # а затем в цикле while переопределяем на минимально возможное
    max_value = sys.maxsize
    for vertex in unvisited_vertexes:
        shortest_path[vertex] = max_value
    shortest_path[start] = 0

    while unvisited_vertexes:
        # ищем вершину с меньшей оценкой
        current_min_vertex = None
        for vertex in unvisited_vertexes:
            if current_min_vertex == None:
                current_min_vertex = vertex
            elif shortest_path[vertex] < shortest_path[current_min_vertex]:
                current_min_vertex = vertex
        neighbors = vertex_neighbors(graf, current_min_vertex)
        for neighbor in neighbors:
            tentative_value = shortest_path[current_min_vertex] + value(graf, current_min_vertex, neighbor)
            if tentative_value < shortest_path[neighbor]:
                shortest_path[neighbor] = tentative_value
                previous_vertex[neighbor] = current_min_vertex
        unvisited_vertexes.remove(current_min_vertex)
    return previous_vertex, shortest_path


def print_result(previous_vertex, shortest_path,
                 start_vertex, target_vertex):
    path = []
    vertex = target_vertex
    while vertex != start_vertex:
        path.append(vertex)
        vertex = previous_vertex[vertex]
    path.append(start_vertex)
    return "->".join(reversed(path)), shortest_path[target_vertex]


previous_vertex, shortest_path = dijkstra(graf, start)
pathway, track = print_result(previous_vertex, shortest_path, start, end)

print(pathway)
print(track)

f = open('output.txt', 'w')
f.write('{pathway}\n{track}'.format(pathway=pathway, track=track))
f.close()
```

```python
def count_components(graph, n_vertices):
    visited = [False] * n_vertices
    count = 0

    for i in range(n_vertices):
        if not visited[i]:
            count += 1
            dfs(i, visited, graph)
    return count


def dfs(v, visited, graph):
    visited[v] = True
    for u in graph[v]:
        if not visited[u]:
            dfs(u, visited, graph)


n = int(input())
m = int(input())
gr = [list(map(int, input().split())) for _ in range(m)]

d_gr = {i: [] for i in range(n)}
for item in gr:
    d_gr[item[0]].append(item[1])
    d_gr[item[1]].append(item[0])

print(count_components(d_gr, n))
```
