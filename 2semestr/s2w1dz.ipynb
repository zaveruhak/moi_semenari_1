D! N.B. Также легко заметить следующий факт – в любом графе число вершин нечетной степени – четно.

докажем по индукции. пусть есть некоторый граф, состоящий из m > 1 вершин(без ребер). пока у него нет ребер, число вершин нечетной степени равно 0 - чётно. добавим ребро, тогда у него станет две вершины степени 1 - условие выполняется. пусть после добавления k ребер число вершин нечётной степени чётно. рассмотрим k + 1 ребро: если оно инцидентно двум (в прошлом) вершинам чётной степени, то вершин с нечётной степенью стало на две больше => чётность сохранилась; если он инцидентно двум(в прошлом) вершинам нечётной степени, то вершин с нечётной степенью стало на две меньше => чётность сохранилась; если же оно инцидентно вершинам чётной и нечётно степени, то их чётности "поменяются" => сохранится кол-во вершин нечётной степени => чётность не изменится. чтд

```python
from math import inf

# n_ver = int(input())
# gr = [tuple(map(int, input().split())) for _ in range(n_ver)]
n_ver = 6
gr = [(0, 1), (1, 2), (2, 3), (3, 5), (4, 1)]

dict_gr = {i: set() for i in range(n_ver)}
for item in gr:
    dict_gr[item[0]].add(item[1])
    dict_gr[item[1]].add(item[0])

way = [inf for _ in range(n_ver)]
way[0] = 1
ne_pos = list(range(n_ver))
while ne_pos and any([way[i] != inf for i in ne_pos]):
    ver = min([(way[i], i) for i in ne_pos])[1]
    ne_pos.remove(ver)
    if ver in dict_gr:
        for item in dict_gr[ver]:
            way[item] = 1

print(inf not in way)
```

```python
n_ver = 5
gr = [(0, 1, 3), (0, 4, 1), (1, 2, 1), (1, 3, 3), (1, 4, 1), (4, 2, 2), (4, 3, 1)]
beg, end = 0, 2

d_gr = {}
for item in gr:
    if item[0] in d_gr:
        d_gr[item[0]].append((item[1], item[2]))
    else:
        d_gr[item[0]] = [(item[1], item[2])]

dist = [10 ** 10 if i != beg else 0 for i in range(n_ver)]
ne_pos = list(range(n_ver))
while ne_pos:
    ver = min([(dist[i], i) for i in ne_pos])[1]
    ne_pos.remove(ver)
    if ver in d_gr:
        for item in d_gr[ver]:
            if dist[ver] + item[1] < dist[item[0]]:
                dist[item[0]] = dist[ver] + item[1]

print(dist[end])
```

```python
vvod = ['ANT', 'OSTRICH', 'DEER', 'TURKEY', 'KANGAROO', 'TIGER', 'RABBIT', 'RAT', 'TOAD', 'YAK', 'HYENA']

gr = [(i[0], i[-1]) for i in vvod]
beg = gr[0]
ends = []


def e_next(ver, graph):
    graph.remove(ver)
    if len(graph) <= 0:
        ends.append(ver[1])
        return
    for elem in filter(lambda x: x[0] == ver[1], graph):
        e_next(elem, graph[:])
    return


e_next(beg, gr)
print(beg[0] in ends)
```
