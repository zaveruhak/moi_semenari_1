```python
s = input().split('student_')
t = {i[:3]:i[3:] for i in s[1:]}
m = max(t.values())
for i in t:
    if t.get(i) == m:
        print(i)
```

```python
s = input()

if len(s) < 4:
    print(s)
else:
    if [s[i].isupper() for i in range(4)].count(False) == 1:
        print(s.upper())
    else:
        print(s)
```

```python
def zak(teg, stroka):
    if teg not in ('a', 'abbr', 'b', 'body', 'caption', 'cite', 'code', 'div', 'form', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'header', 'i', 's'):
        print("Введён неверный тег")
    else:
        print(f"<{teg}>{stroka}</{teg}>")

zak('i', 'Python')
```

```python
a = input()

if len(a) <= 2:
    print(ord(a[0]))  
elif len(a) < 10:
    print(sum(map(ord, (a[0], a[len(a) // 2 - 1 ^ (len(a) % 2)], a[-1]))))         
else:
    print(ord(a[-1]))
```
