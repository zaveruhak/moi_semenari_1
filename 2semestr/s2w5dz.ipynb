```python
n = int(input())
coordinates = [tuple(map(int, input().split())) for _ in range(n)]
graph = [[10000000] * n] * n
for i in range(n):
    for j in range(n):
        graph[i][j] = ((coordinates[i][0] - coordinates[j][0]) ** 2 + (
                    coordinates[i][1] - coordinates[j][1]) ** 2) ** 0.5

dist = [10 ** 10 if i != 0 else 0 for i in range(n)]
ne_pos = list(range(n))
while ne_pos:
    ver = min([(dist[i], i) for i in ne_pos])[1]
    ne_pos.remove(ver)
    for item in range(n):
        if dist[item] > max(dist[ver], graph[ver][item]):
            dist[item] = max(dist[ver], graph[ver][item])
print(f'{dist[1]:.3f}')
```

```python
ans = []
for _ in range(int(input())):
    n, m = map(int, input().split())
    coordinates = [tuple(map(int, input().split())) for _ in range(m)]
    graph = dist = [[float('inf')] * n for _ in range(n)]
    for i in range(m):
        graph[coordinates[i][0]][coordinates[i][1]] = coordinates[i][2]

    dist = [[float('inf')] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            if i == j:
                dist[i][j] = 0
            else:
                dist[i][j] = graph[i][j]

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]

    f = False
    for k in range(n):
        if dist[k][k] < 0:
            f = True
            break
    if f:
        ans.append('возможно')
    else:
        ans.append('не возможно')
for i in ans:
    print(i)
```
