```python
# with open('input.txt') as f:
#     s = [i.split() for i in f.readlines()]
# n = int(s[0][0])
# name = s[1][0]
# cubes = []
# for i in range(n):
#     cubes.append(s[i + 2][0])

n = 5
name = 'HELEN'
m = len(name)
cubes = ['ABCDEF', 'GHIJKL', 'MNOPQL', 'STUVWN', 'EIUOZK']

graph = [[] for _ in range(m)]
for i in range(m):
    ch = name[i]
    for j in range(n):
        if ch in cubes[j]:
            graph[i].append(j)

match = [-1] * n


def dfs(u):
    for v in graph[u]:
        if not used[v]:
            used[v] = True
            if match[v] == -1 or dfs(match[v]):
                match[v] = u
                return True
    return False


f = True
for i in range(m):
    used = [False] * n
    if not dfs(i):
        f = False
        break

ans = [0] * m
if f:
    for j in range(n):
        if match[j] != -1:
            ans[match[j]] = j + 1

# with open('output.txt', 'w') as t:
#     if f:
#         t.write("YES\n")
#         t.write(' '.join(map(str, ans)))
#     else:
#         t.write("NO")

if f:
    print('YES')
    print(*ans)
else:
    print('NO')
```
