первый номер

```python
n = int(input())
A = list(map(int, input().split()))
k = int(input())
z = [tuple(map(int, input().split())) for _ in range(k)]


def build_tree_1(a):
    tree = {}
    for i in range(n):
        tree[n + i] = a[i]
    for i in range(n - 1, 0, -1):
        tree[i] = max(tree[2 * i], tree[2 * i + 1])
    return tree


def max_tree(tree, l, r):
    _max = 0
    l += n
    r += n
    while l < r:
        if l & 1:
            _max = max(tree[l], _max)
            l += 1
        if r & 1:
            r -= 1
            _max = max(tree[r], _max)
        l //= 2
        r //= 2
    return _max


tr = build_tree_1(A)
for left, right in z:
    print(max_tree(tr, left, right))
```

второй номер

```python
def gcd1(a, b):
    while b:
        a, b = b, a % b
    return a


n = int(input())
A = list(map(int, input().split()))
k = int(input())
z = [tuple(map(int, input().split())) for _ in range(k)]


def build_tree_2(a):
    tree = {}
    for i in range(n):
        tree[n + i] = a[i]
    for i in range(n - 1, 0, -1):
        tree[i] = gcd1(tree[2 * i], tree[2 * i + 1])
    return tree


def gcd_tree(tree, l, r):
    _gcd = 0
    l += n
    r += n
    while l < r:
        if l & 1:
            _gcd = gcd1(tree[l], _gcd)
            l += 1
        if r & 1:
            r -= 1
            _gcd = gcd1(tree[r], _gcd)
        l //= 2
        r //= 2
    return _gcd


tr = build_tree_2(A)
for left, right in z:
    print(gcd_tree(tr, left, right))
```

третий номер

```python
n = int(input())
A = list(map(int, input().split()))
k = int(input())
z = [tuple(map(int, input().split())) for _ in range(k)]
zeros = [1 if x == 0 else 0 for x in A]
tree = [0] * (2 * n)
for i in range(n):
    tree[n + i] = zeros[i]
for i in range(n - 1, 0, -1):
    tree[i] = tree[2 * i] + tree[2 * i + 1]


def sum_tree(l, r):
    _sum = 0
    l += n
    r += n
    while l < r:
        if l & 1:
            _sum += tree[l]
            l += 1
        if r & 1:
            r -= 1
            _sum += tree[r]
        l //= 2
        r //= 2
    return _sum


def do_null(l, r, k):
    l0 = l - 1
    r0 = r - 1
    ans = -1
    while l0 <= r0:
        m = (l0 + r0) // 2
        cnt = sum_tree(l0, m + 1)
        if cnt >= k:
            ans = m
            r0 = m - 1
        else:
            l0 = m + 1
    return ans + 1


for left, right, k in z:
    print(do_null(left, right, k))
# выводит номер позиции нуля от абсолютного начала
```
