# Упражнение 1.

```python
def p(x):
    return(all([x % i != 0 for i in range(2, int(x ** 0.5) + 1)]))


def raz(a):
    if a <= 0:
        raise ValueError
    if a == 1:
        return []
    for i in range(2, int(a ** 0.5) + 1):
        if a % i == 0 and p(i):
            return [i] + raz(a // i)
    return [a]
```

```python
%%unittest_main


class TestOrdinaryNumber(unittest.TestCase):
    def test_raz(self):
        self.assertEqual(raz(6), [2, 3], "should be [2, 3]")


class TestPrimeNumber(unittest.TestCase):
    def test_raz(self):
        self.assertEqual(raz(13), [13], "should be [13]")


class TestIncorrectArgs(unittest.TestCase):
    def test_raz(self):
        with self.assertRaises(TypeError):
            raz("2")


class TestNegative(unittest.TestCase):
    def test_raz(self):
        with self.assertRaises(ValueError):
            raz(-6)


class TestZero(unittest.TestCase):
    def test_raz(self):
        with self.assertRaises(ValueError):
            raz(0)
```

# Упражнение 2.

```python
import numpy as np


def least_squares(x, y):
    if not (isinstance(x, np.ndarray) and isinstance(y, np.ndarray)) or not (
            (x.dtype == 'float64' and y.dtype == 'float64') or (x.dtype == 'int64' and y.dtype == 'int64')):
        raise TypeError
    if len(x) != len(y):
        raise ValueError

    k = (np.mean(x * y) - np.mean(x) * np.mean(y)) / (np.mean(x * x) - np.mean(x) ** 2)
    b = np.mean(y) - k * np.mean(x)
    return k, b
```

```python
%%unittest_main


class TestOrdinaryNumbers(unittest.TestCase):
    def test_least_squares(self):
        self.assertEqual(least_squares(np.array([6, 4]), np.array([5, 4])), (np.float64(0.5), np.float64(2.0)),
                         "should be (np.float64(0.5), np.float64(2.0))")


class TestUnevenLists(unittest.TestCase):
    def test_least_squares(self):
        with self.assertRaises(ValueError):
            least_squares(np.array([6]), np.array([5, 4]))


class TestIncorrectType(unittest.TestCase):
    def test_least_squares(self):
        with self.assertRaises(TypeError):
            least_squares(np.array(5), np.array(''))
```

# Упражнение 3.

```python
import random


def quick_sorted(e):
    for v in e:
        if not isinstance(v, (int, float)):
            raise TypeError

    def quick_sort(a, l, r):
        if l >= r:
            return
        else:
            q = random.choice(a[l:r + 1])
            i = l
            j = r
            while i <= j:
                while a[i] < q:
                    i += 1
                while a[j] > q:
                    j -= 1
                if i <= j:
                    a[i], a[j] = a[j], a[i]
                    i += 1
                    j -= 1
                    quick_sort(a, l, j)
                    quick_sort(a, i, r)

    quick_sort(e, 0, len(e) - 1)
    return e
```

```python
%%unittest_main


class TestOrdinaryNumbers(unittest.TestCase):
    def test_quick_sorted(self):
        self.assertEqual(quick_sorted([-3, 9, 4, 1, -2, 3]), [-3, -2, 1, 3, 4, 9], "should be [-3, -2, 1, 3, 4, 9]")


class TestIncorrectArgs(unittest.TestCase):
    def test_quick_sorted(self):
        with self.assertRaises(TypeError):
            quick_sorted("2", 3, 6, 4, -8, 10)
```

# Упражнение 4.

```python
def rashifrovka(key_vn, text_vn):
    class Caesar:
        alphabet = "яюэьыъщшчцхфутсрпонмлкйизжёедгвба"
        if not isinstance(key_vn, int):
            raise TypeError

        def __init__(self, key):
            self._encode = dict()
            for i in range(len(self.alphabet)):
                letter = self.alphabet[i]
                encoded = self.alphabet[(i + key) % len(self.alphabet)]
                self._encode[letter] = encoded
                self._encode[letter.upper()] = encoded.upper()
            self._decode = dict()

            for i in range(len(self.alphabet)):
                letter = self.alphabet[i]
                self._decode[self._encode[letter]] = letter
                self._decode[self._encode[letter.upper()]] = letter.upper()

        def encode(self, text):
            return ''.join([self._encode.get(char, char) for char in text])

        def decode(self, text):
            return ''.join([self._decode.get(char, char) for char in text])

    cipher = Caesar(key_vn)
    return cipher.decode(text_vn)
```

```python
%%unittest_main


class TestOrdinary(unittest.TestCase):
    def test_rashifrovka(self):
        self.assertEqual(rashifrovka(14, 'Вгбхгтяяё цюс йтдебеабхб татюыът дючцёче атвыдтео дтябдебсечюоаб. Ёдвчзбф!'), 'Программу для частотного анализа следует написать самостоятельно. Успехов!')

class TestIncorrectArgs(unittest.TestCase):
    def test_rashifrovka(self):
        with self.assertRaises(TypeError):
            rashifrovka("2", 'йцукенгшщзхъфывапролджэячсмитьбю')
```
