```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

linear_data10 = pd.read_csv('111data10.csv', index_col=0)
linear_data20 = pd.read_csv('111data20.csv', index_col=0)
linear_data30 = pd.read_csv('111data30.csv', index_col=0)
linear_data50 = pd.read_csv('111data50.csv', index_col=0)
```

```python
def least_squares(x, y):
    coeff = np.mean(x*y) / np.mean(x*x)
    err = 1 / np.sqrt(len(x)) * np.sqrt((y*y).mean() / (x*x).mean() - coeff ** 2)
    return coeff, err
```

```python
coeff10, err10 = least_squares(linear_data10['X10'], linear_data10['Y10'])
coeff20, err20 = least_squares(linear_data20['X20'], linear_data20['Y20'])
coeff30, err30 = least_squares(linear_data30['X30'], linear_data30['Y30'])
coeff50, err50 = least_squares(linear_data50['X50'], linear_data50['Y50'])

print(coeff10, err10, '-- R & погрешность для 10см')
print(coeff20, err20, '-- R & погрешность для 20см')
print(coeff30, err30, '-- R & погрешность для 30см')
print(coeff50, err50, '-- R & погрешность для 50см')
```

```python
tab, ax = plt.subplots()
ax.set_xlim(0, 400)
ax.set_ylim(0, 1300)
plt.xlabel('I, мА')
plt.ylabel('V, мВ')

plt.scatter(linear_data10['X10'], linear_data10['Y10'], color='tab:red')
plt.axline((0, 0), slope=coeff10, color='red', label='10cm')

plt.scatter(linear_data20['X20'], linear_data20['Y20'], color='tab:blue')
plt.axline((0, 0), slope=coeff20, color='blue', label='20cm')

plt.scatter(linear_data30['X30'], linear_data30['Y30'], color='tab:green')
plt.axline((0, 0), slope=coeff30, color='green', label='30cm')

plt.scatter(linear_data50['X50'], linear_data50['Y50'], color='tab:orange')
plt.axline((0, 0), slope=coeff50, color='orange', label='50cm')

plt.legend()
plt.grid()
plt.show()
```
